```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/student-mat.csv", sep=";")
df.head()
```

```python
df.shape
df.columns
df.info()
df.describe()
```

```python
df["pass"] = df["G3"].apply(lambda x: 1 if x >= 10 else 0)
df["pass"].value_counts()
```

```python
sns.countplot(x="pass", data=df)
plt.title("Pass vs Fail Distribution")
plt.show()
```

```python
sns.boxplot(x="pass", y="studytime", data=df)
plt.title("Study Time vs Pass/Fail")
plt.show()
```

```python
sns.boxplot(x="pass", y="absences", data=df)
plt.title("Absences vs Pass/Fail")
plt.show()
```

Conclusion:
This dataset contains academic and lifestyle features of students.
Students with higher study time and lower absences tend to pass.
A binary target variable (pass/fail) has been created for prediction.
